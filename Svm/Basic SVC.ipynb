## Support Vector Machine (SVM) Classifier (SVC)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Creating a Dataset
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
## Predictions
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
svc.intercept_, svc.coef_, svc.support_vectors_
```

```python
## Cluster per class = 2
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
svc = SVC(kernel='rbf')
```

```python
svc.fit(X_train, y_train)
```

```python
## Predictions
y_pred = svc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
svc = SVC(kernel='poly')
svc.fit(X_train, y_train)
```

```python
## Predictions
y_pred = svc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
svc = SVC(kernel='sigmoid')
svc.fit(X_train, y_train)
```

```python
## Predictions
y_pred = svc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
svc.intercept_
```

### Hyperparameter Tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']} ## Directly selecting the kernel as rbf as it is the best kernel for this dataset
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3, cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
## Predictions
y_pred = grid.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
