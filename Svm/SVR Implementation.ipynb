```python
## Independent and Dependent Features
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
X
```

```python
y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
## Feature Encoding (Label Encoding and One Hot Encoding)
from sklearn.preprocessing import LabelEncoder
```

```python
lab_1 = LabelEncoder()
lab_2 = LabelEncoder()
lab_3 = LabelEncoder()
```

```python
X_train['sex'] = lab_1.fit_transform(X_train['sex'])
X_train['smoker'] = lab_2.fit_transform(X_train['smoker'])
X_train['time'] = lab_3.fit_transform(X_train['time'])
```

```python
X_train
```

```python
X_test['sex'] = lab_1.transform(X_test['sex'])
X_test['smoker'] = lab_2.transform(X_test['smoker'])
X_test['time'] = lab_3.transform(X_test['time'])
```

```python
X_test
```

```python
## One Hot Encoding -- Column Transformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(drop='first'), [3])], remainder='passthrough')
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
X_train = ct.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = ct.transform(X_test)
```

```python
X_test
```

```python
## SVR immplementation
from sklearn.svm import SVR
```

```python
svr = SVR()
```

```python
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))
```

```python
## Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVR(), param_grid, refit=True, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))
```
