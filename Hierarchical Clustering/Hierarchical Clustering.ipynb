## Hierarchical Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
df= pd.DataFrame(data= iris['data'], columns= iris['feature_names'])
```

```python
df.head()
```

```python
df.tail()
```

```python
## Standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled_data = scaler.fit_transform(df)
```

```python
X_scaled_data
```

```python
plt.scatter(X_scaled_data[:,0], X_scaled_data[:,1], c= iris['target'])
```

```python
## Apply PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=4)
pca.fit_transform(X_scaled_data)
pca.explained_variance_ratio_.cumsum()
```

```python
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X_scaled_data)
```

```python
X_pca
```

```python
plt.scatter(X_pca[:,0], X_pca[:,1])
```

```python
## To confirm the variance captured by the two components
plt.scatter(X_pca[:,0], X_pca[:,1], c= iris['target'])
```

```python
## Agglomerative Clustering
## Constructing the Dendrogram
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(25, 20))
plt.title('Dendrogram')
sch.dendrogram(sch.linkage(X_pca, method='ward'))
plt.xlabel('Iris')
plt.ylabel('Euclidean distances')
plt.axhline(y=8, color='r', linestyle='-')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
```

```python
cluster.fit(X_pca)
```

```python
cluster.labels_
```

```python
plt.scatter(X_pca[:,0], X_pca[:,1], c= cluster.labels_)
```
