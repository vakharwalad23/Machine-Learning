### 1. Nominal/OHE Encoding (One Hot Encoding)
##### This technique is used to represent categorical data as a numerical data, which is more suitable for machine learning algos.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## DataFrame
df = pd.DataFrame({
    'color':['red', 'green', 'blue', 'green', 'blue', 'red']
})
```

```python
df.head()
```

```python
# Create an instance of OHE
ohe = OneHotEncoder()
```

```python
## Fit and transform
encoded=ohe.fit_transform(df[['color']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded,columns=ohe.get_feature_names_out())
```

```python
encoded_df
```

```python
pd.concat([df, encoded_df],axis=1)
```

```python
## Example
import seaborn as sns
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.tail()
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['sex', 'smoker', 'day', 'time']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoded_df.head()
```

```python
pd.concat([df,encoded_df], axis=1)
```

### 2. Label Encoding
##### This technique is used to represent categorical data as a numerical data. It involves assigning a unique label to each category in the variable.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({
    'color':['red', 'green', 'blue', 'green', 'blue', 'red']
})
```

```python
lble = LabelEncoder()
```

```python
lble.fit_transform(df[['color']])
```

```python
lble.transform([['red']])
```

```python
lble.transform([['green']])
```

```python
lble.transform([['blue']])
```

### 3. Ordinal Encoding
##### This technique is used to represent categorical data as a numerical data which has intrinsic order or ranking.

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({
    'size':['s', 'm', 'l', 'm', 's', 'l']
})
```

```python
oe = OrdinalEncoder(categories=[['s', 'm', 'l']])
```

```python
oe.fit_transform(df[['size']])
```

```python
oe.transform([['l']])
```

```python
oe.transform([['s']])
```

### 4.Target Guided Ordinal Encoding
##### The main idea behind the target encoder is to encode the categories by replacing them for a measurement of the effect they might have on the target.

```python
df = pd.DataFrame({
    'city': ['London', 'Paris', 'NY', 'Tokyo', 'NY', 'London'],
    'price': [100, 120, 300, 250, 240, 450]
})
```

```python
encoded = df.groupby('city')['price'].mean().to_dict()
```

```python
df['city_encoded'] = df['city'].map(encoded)
```

```python
df[['price', 'city_encoded']]
```
