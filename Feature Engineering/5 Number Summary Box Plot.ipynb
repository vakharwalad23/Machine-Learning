### 5 Number summary
1. Minimum
2. Maximum
3. Median
4. Q1
5. Q3
6. IQR

```python
import numpy as np
```

```python
list_marks = [45, 32, 56, 75, 89, 54, 32, 89, 90, 87, 67, 54, 45, 98, 99, 67, 74]

min,q1,median,q3,max = np.quantile(list_marks, [0,0.25,0.50,0.75,1.0])
```

```python
min,q1,median,q3,max
```

```python
iqr = q3-q1
iqr
```

```python
lower_fence = q1-1.5*(iqr)
higher_fence = q3+1.5*(iqr)
```

```python
lower_fence,higher_fence
```
