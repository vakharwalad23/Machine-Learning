# TF-IDF

TF-IDF (Term Frequency-Inverse Document Frequency) is a statistical measure used to evaluate the importance of a word in a document relative to a collection of documents (corpus). It is commonly used in information retrieval and text mining. TF-IDF is calculated by multiplying two metrics: Term Frequency (TF) and Inverse Document Frequency (IDF). TF-IDF vectors are often used in machine learning and natural language processing tasks to represent text data.

```python
import pandas as pd

df = pd.read_csv("SMSSpamCollection.csv", sep="\t", names=["label", "message"])
```

```python
df.head()
```

```python
df.tail()
```

```python
# Data cleaning and preprocessing
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(df)):
    review = re.sub("[^a-zA-Z]", " ", df["message"][i]) # keep only letters
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words("english")]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# TF-IDF Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=500)
```

```python
X = vectorizer.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
import numpy as np

np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float=lambda x: "%.3f" % x))
```

```python
X
```

```python
vectorizer.vocabulary_
```

```python
# TF-IDF with N-grams
vectorizer_ngram = TfidfVectorizer(max_features=1500, ngram_range=(1, 2))
X_ngram = vectorizer_ngram.fit_transform(corpus).toarray()
```

```python
X_ngram.shape
```

```python
X_ngram
```

```python
vectorizer_ngram.vocabulary_
```

```python
## Dependent feature
y = pd.get_dummies(df["label"], drop_first=True)
y
```

```python
# Convert y: True to 1 and False to 0
y = y.astype(int)
```

```python
y
```

```python
y.values.shape
```

```python
# Train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_ngram, y, test_size=0.2, random_state=42)
```

```python
# Naive Bayes Classifier works best with sparse matrices
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_classifier = MultinomialNB()
spam_classifier.fit(X_train, y_train.values.ravel()) # to avoid warning
```

```python
y_pred = spam_classifier.predict(X_test)
```

```python
y_pred
```

```python
# Accuracy matrix
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
