# Bag of Words

Bag of Words (BoW) is a simple and commonly used technique in natural language processing (NLP) for representing text data. It involves converting text into a fixed-length vector by counting the occurrences of each word in a predefined vocabulary.

```python
import pandas as pd

df = pd.read_csv("SMSSpamCollection.csv", sep="\t", names=["label", "message"])
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(df)):
    review = re.sub("[^a-zA-Z]", " ", df["message"][i]) # keep only letters
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words("english")]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# create bag of words model
from sklearn.feature_extraction.text import CountVectorizer

# For Binary Bag of Words enable binary=True or for Normal Bag of Words disable binary or set it to False
vectorizer = CountVectorizer(
    max_features=2500,
    binary=True,
    ngram_range=(1, 2)
)
```

```python
## Independent features
X = vectorizer.fit_transform(corpus).toarray()
```

```python
X
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

```python
## Dependent feature
y = pd.get_dummies(df["label"], drop_first=True)
```

```python
y
```

```python
# Convert y: True to 1 and False to 0
y = y.astype(int)
```

```python
y
```

```python
y.values.shape
```

```python
# Train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Naive Bayes Classifier works best with sparse matrices
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_classifier = MultinomialNB()
spam_classifier.fit(X_train, y_train.values.ravel()) # to avoid warning
```

```python
y_pred = spam_classifier.predict(X_test)
```

```python
y_pred
```

```python
# Accuracy matrix
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
