## Logistic Regression (OVO - One vs One and OVR - One vs Rest)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Making dataset

```python
from sklearn.datasets import make_classification
```

```python
## create the dataset
X, y = make_classification(n_samples=10000, n_features=10, n_classes=2, random_state=42)
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=42)
```

### Implementing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_reg = LogisticRegression()
```

```python
logistic_reg.fit(X_train, y_train)
```

```python
y_pred = logistic_reg.predict(X_test)
```

```python
print(y_pred)
```

```python
y_proba = logistic_reg.predict_proba(X_test)
```

```python
print(y_proba)
```

### Performance Metrics

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
score
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
cr = classification_report(y_test, y_pred)
print(cr)
```

## Hyperparameter Tuning and Cross Validation

```python
logistic_reg = LogisticRegression()
```

```python
penalty = ['l1', 'l2', 'elasticnet']
c_vals = [0.1,0.01,1.0,10.0,100.0]
solvers = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
hyperparameters = dict(penalty = penalty, C = c_vals, solver = solvers)
```

```python
hyperparameters
```

#### Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
```

```python
cv = StratifiedKFold()
grid = GridSearchCV(estimator=logistic_reg, param_grid=hyperparameters, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
## Scoring
score = accuracy_score(y_test, y_pred)
score
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
cr = classification_report(y_test, y_pred)
print(cr)
```
