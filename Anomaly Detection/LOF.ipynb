## LOF - Local Outlier Factor Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Healthcare Data.csv')
```

```python
df.head()
```

```python
plt.scatter(df.iloc[:,0], df.iloc[:,1], c='blue')
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof_outlier = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
```

```python
outier_scores = lof_outlier.fit_predict(df)
```

```python
outier_scores
```

```python
abn_idx = np.where(outier_scores==-1)
```

```python
abn_idx
```

```python
X = df.values
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1], c='blue')
plt.scatter(X[abn_idx,0], X[abn_idx,1], edgecolors='red')
plt.legend(["inliers", "outliers"])
```
